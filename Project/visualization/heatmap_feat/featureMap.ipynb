```python
# import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
# read in data
def heatmap_plot(data):
    cols = range(1,100)
    df = pd.read_csv(data, header=None, skiprows=[0], delim_whitespace=True, usecols=cols)
    sns.heatmap(df, cmap='RdYlGn_r', annot=False, xticklabels=False, yticklabels=False)
```

```python
# vgg_final_layer data
data_vgg_final = '../../data/features/vgg_ssv_final_layer/Train_vgg/ID111_vid4_image.ssv'
heatmap_plot(data_vgg_final)
```

```python
# vgg_final_layer data
data_vggFace2 = '../../data/features/vggFace2_layer/Train_vggFace2_feature/ID111_vid4_image.ssv'
heatmap_plot(data_vggFace2)
```

```python
# vgg_final_layer data
data_raw = '../../data/features/raw_img/ID111_vid4_image.ssv'
heatmap_plot(data_raw)
```

```python
# extracting face features
data_face = '../../data/features/ID111_vid4_face.ssv'
heatmap_plot(data_face)
```
